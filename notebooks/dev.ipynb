```python
R = []
for w in pd.read_csv("../data/mtb_trna_synthetases_bosch_2021_fig5.csv").values:
    n = w[0]
    uniprot_ac = w[1]
    for v in dp[["Entry", "Protein names", "Gene Names", "EC number", "Sequence"]].iterrows():
        if v[1]["Entry"] == uniprot_ac:
            R += [[n] + list(v[1])]
            break
df = pd.DataFrame(R, columns = ["gene_name_in_bosch_2021", "uniprot_ac", "protein_names", "gene_names", "ec_number", "sequence"])
```

```python
df.to_csv("mtb_rna_synthetases_bosch_2021_fig5_proteome.csv", index=False)
```

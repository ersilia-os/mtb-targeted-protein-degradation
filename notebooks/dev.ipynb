```python
import requests
import os

def fetch_pdb_from_uniprot(uniprot_ac, output_dir="pdb_files"):
    # Step 1: Map UniProt AC to PDB IDs using UniProt API
    uniprot_url = f"https://rest.uniprot.org/uniprotkb/{uniprot_ac}.json"
    response = requests.get(uniprot_url)
    
    if response.status_code != 200:
        print(f"Error fetching data for UniProt AC: {uniprot_ac}")
        return

    # Parse the JSON response to find PDB IDs
    data = response.json()
    pdb_ids = []
    if "dbReferences" in data:
        for ref in data["dbReferences"]:
            if ref["type"] == "PDB":
                pdb_ids.append(ref["id"])

    if not pdb_ids:
        print(f"No PDB IDs found for UniProt AC: {uniprot_ac}")
        return

    # Step 2: Fetch PDB files using RCSB PDB
    os.makedirs(output_dir, exist_ok=True)
    for pdb_id in pdb_ids:
        pdb_url = f"https://files.rcsb.org/download/{pdb_id}.pdb"
        pdb_response = requests.get(pdb_url)

        if pdb_response.status_code == 200:
            with open(f"{output_dir}/{pdb_id}.pdb", "w") as pdb_file:
                pdb_file.write(pdb_response.text)
            print(f"PDB structure {pdb_id} downloaded successfully.")
        else:
            print(f"Failed to download PDB structure {pdb_id}.")

# Example usage
fetch_pdb_from_uniprot("P9WFT1")  # Example UniProt AC for hemoglobin subunit alpha
```
